```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

### Checking for missing values

```python
#  Check for missing values
missing_values = data.isnull().sum()
```

```python
missing_values
```

### Label Encoding

```python
# Label Encoding for the target variable
label_encoder = LabelEncoder()
data['letter'] = label_encoder.fit_transform(data['letter'])
```

### Feature Scaling

```python
# Feature Scaling
scaler = StandardScaler()
features = data.drop(columns='letter')
scaled_features = scaler.fit_transform(features)
```

```python
data
```

### Train Test split

```python
#  Train-test Split
X_train, X_test, y_train, y_test = train_test_split(
    scaled_features, data['letter'], test_size=0.2, random_state=42)
```

```python
# shapes of train and test datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

### Model Building

```python
#  Build the Model
model = Sequential()
```

### Input layer

```python
# Input layer
model.add(Dense(128, input_shape=(16,), activation='relu'))
```

##### Here i use relu as activation function

### hidden layers

```python
# Hidden layers
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
```

### Output Layer

```python
# Output layer
model.add(Dense(26, activation='softmax'))  # 26 classes for 26 letters
```

```python
# Compile the Model
model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
#  Train the Model
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping]
)
```

```python
# Evaluate the Model
evaluation = model.evaluate(X_test, y_test)
print(f'Test Loss: {evaluation[0]}, Test Accuracy: {evaluation[1]}')
```

```python
# Hyperparameter Tuning
def build_model(hp):
    model = Sequential()
    model.add(Dense(
        units=hp.Int('units_input', min_value=32, max_value=512, step=32),
        activation='relu',
        input_shape=(16,)
    ))
    for i in range(hp.Int('num_layers', 1, 4)):
        model.add(Dense(
            units=hp.Int(f'units_{i}', min_value=32, max_value=512, step=32),
            activation='relu'
        ))
        model.add(Dropout(rate=hp.Float(f'dropout_{i}', 0.0, 0.5, step=0.1)))
    model.add(Dense(26, activation='softmax'))
    model.compile(
        optimizer=Adam(learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
# Set up the tuner
import kerastuner as kt
tuner = kt.RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='alphabet_classification'
)
```

```python
# Perform hyperparameter search
tuner.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
# Get the optimal hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
```

```python
# Build the model with the optimal hyperparameters and train it
model = tuner.hypermodel.build(best_hps)
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=32, callbacks=[early_stopping])
```

```python
# Evaluate the best model
evaluation = model.evaluate(X_test, y_test)
print(f'Test Loss: {evaluation[0]}, Test Accuracy: {evaluation[1]}')
```
