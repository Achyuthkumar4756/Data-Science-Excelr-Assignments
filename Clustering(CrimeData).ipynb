# Scaling Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df.iloc[:,1:])
```

```python
scaled_df =pd.DataFrame(scaled_df)
```

# Applying kmeans clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0) 

kmeans.fit(scaled_df)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

# Elbow method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11),wcss,color = 'red')
plt.scatter(range(1,11),wcss,color='green')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

By the above plot, we can see that there is a kink at k=4

```python
wcss
```

# Applying Hierarichal Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

# Normalizing data

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:])
df_norm.head()
```

# Dendrogram

```python
# create dendrogram
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='single'))
```

```python
fig = plt.figure(figsize=(15,8))
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='average'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(15,8))
# Compute the clusters
Z = linkage(scaled_df, method='ward')

# Plot dendrogram
dendrogram(Z)
plt.show()
```

```python
crimes = pd.concat([df,Clusters],axis=1)
crimes
```

```python
crimes['Clusters'].value_counts()
```

# Applying DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
# Create DBSCAN instance
dbscan = DBSCAN(eps=0.98, min_samples=3)
```

```python
# Fit the model to your data
dbscan.fit(scaled_df)
```

```python
# Get cluster labels
labels = dbscan.labels_
```

```python
labels
```

```python
Cluster_db= pd.DataFrame(dbscan.labels_,columns=['Cluster ID'])
Cluster_db.value_counts()
```
